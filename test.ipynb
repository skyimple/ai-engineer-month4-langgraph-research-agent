```python
import os
from langchain_openai import ChatOpenAI  # 注意：这里改用 ChatOpenAI
from langchain_core.messages import HumanMessage
from dotenv import load_dotenv

# 加载环境变量
load_dotenv("config.env")

api_key = os.getenv("DASHSCOPE_API_KEY")
if not api_key:
    raise ValueError("未找到 DASHSCOPE_API_KEY")

print("正在使用 ChatOpenAI 兼容模式调用 Qwen3.5-Plus...")

# 初始化模型
llm = ChatOpenAI(
    model="qwen3.5-plus",  # 模型名称
    temperature=0.7,
    max_tokens=2048,
    streaming=False,
    api_key=api_key,
    # 【关键修复】显式指定阿里云的 OpenAI 兼容端点
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1"
)

# 简单测试
try:
    response = llm.invoke([HumanMessage(content="你好，你使用的qwen的什么版本")])
    print("\n--- 模型回答 ---")
    print(response.content)
    print("----------------")
except Exception as e:
    print(f"\n❌ 执行错误：{type(e).__name__}: {str(e)}")
    if "401" in str(e):
        print("💡 提示：API Key 可能无效或格式错误。")
    elif "404" in str(e) or "model not found" in str(e).lower():
        print("💡 提示：模型名称 'qwen3.5-plus' 可能有误，请检查阿里云控制台。")
```
